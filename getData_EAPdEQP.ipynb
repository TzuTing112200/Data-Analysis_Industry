```python
import re
import csv
import datetime
import json
```

```python
deviceState = []
with open('EAPdEQP設備狀態.csv', newline='') as csvfile:
    rows = csv.DictReader(csvfile)
    for row in rows:
        deviceState.append((datetime.datetime.strptime(row['交易日期'], "%Y/%m/%d %H:%M:%S"),\
                            row['設備狀態代碼']=="Run"))
print(deviceState[0])
```

```python
sorted(deviceState)
```

```python
jsObj = '\n'.join([json.dumps({str(e[0]):e[1]}) for e in deviceState])
fileObject = open('deviceState_sorted.json', 'w')
fileObject.write(jsObj)
fileObject.close()
```
