```python
import json
from sklearn.metrics import accuracy_score
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score

#有三種錯誤要預測，分別以三個y值存放，0為機台正常，1為錯誤發生；X為機台數據
dataset_X = []
dataset_Y1 = []
dataset_Y2 = []
dataset_Y3 = []
rf = open('record.json', 'r')
for r in rf.readlines():
    dic = json.loads(r)
    for key in dic:
        tmp = dic[key][0]
        del tmp[19] #過濾掉不要的欄位
        del tmp[12]
        del tmp[10]        
        del tmp[8]        
        del tmp[7]        
        del tmp[6]        
        del tmp[5]        
        del tmp[4]
        dataset_X.append(tmp)
        dataset_Y1.append(dic[key][1])        
        dataset_Y2.append(dic[key][2])
        dataset_Y3.append(dic[key][3])
print("X item size:", len(tmp))
print("data sample size: ", len(dataset_X))
rf.close()
```

```python
# test_size 為切分 training data 和 testing data 的比例，0.2 -> 8:2
X_train, X_test, y1_train, y1_test = train_test_split(dataset_X, dataset_Y1, test_size=0.2)
y2_train, y2_test, y3_train, y3_test = train_test_split(dataset_Y2, dataset_Y3, test_size=0.2)

#建立線性回歸模型，create linear regression model
regr1 = linear_model.LinearRegression()
regr2 = linear_model.LinearRegression()
regr3 = linear_model.LinearRegression()

#將訓練的資料放到模型中，Train the model using the training sets
regr1.fit(X_train, y1_train)
regr2.fit(X_train, y2_train)
regr3.fit(X_train, y3_train)
#將剛剛切割出的兩成測試樣本放到模型中預測，use model predict y value
y1_pred = regr1.predict(X_test)
y2_pred = regr2.predict(X_test)
y3_pred = regr3.predict(X_test)
#將線性回歸預測的結果轉為binary的機台狀態，以0.5作為正確與錯誤的邊界
y1_pred_binary=[]
y2_pred_binary=[]
y3_pred_binary=[]
for pre in y1_pred:
    if pre < 0.5:
        y1_pred_binary.append(0)
    else:
        y1_pred_binary.append(1)
for pre in y2_pred:
    if pre < 0.5:
        y2_pred_binary.append(0)
    else:
        y2_pred_binary.append(1)

for pre in y3_pred:
    if pre < 0.5:
        y3_pred_binary.append(0)
    else:
        y3_pred_binary.append(1)


print("y1 acc: ",accuracy_score(y1_test, y1_pred_binary))
print("y2 acc: ",accuracy_score(y2_test, y2_pred_binary))
print("y3 acc: ",accuracy_score(y3_test, y3_pred_binary))
```
