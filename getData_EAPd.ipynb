```python
import re
import csv
import datetime
import json
```

```python
record = {}
i=0
with open('EAPd二銅線生產記錄.csv', newline='') as csvfile:
    rows = csv.DictReader(csvfile)
    for row in rows:
        i = i + 1
        
        data = []
        data.append(float(row['C面積A']))
        data.append(float(row['C面積B']))
        data.append(float(row['S面積A']))
        data.append(float(row['S面積B']))
        
        stac = int(row['錫槽A靶設定電流C'])
        atac = int(row['錫槽A靶實際電流C'])
        data.append(stac)
        if stac == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((atac-stac)/stac)))
        
        stas = int(row['錫槽A靶設定電流S'])
        atas = int(row['錫槽A靶實際電流S'])
        data.append(stas)
        if stas == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((atas-stas)/stas)))
        
        stbc = int(row['錫槽B靶設定電流C'])
        atbc = int(row['錫槽B靶實際電流C'])
        data.append(stbc)
        if stbc == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((atbc-stbc)/stbc)))
        
        stbs = int(row['錫槽B靶設定電流S'])
        atbs = int(row['錫槽B靶實際電流S'])
        data.append(stbs)
        if stbs == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((atbs-stbs)/stbs)))
        
        scac = int(row['銅槽A靶設定電流C'])
        acac = int(row['銅槽A靶實際電流C'])
        data.append(scac)
        if scac == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((acac-scac)/scac)))
        
        scas = int(row['銅槽A靶設定電流S'])
        acas = int(row['銅槽A靶實際電流S'])
        data.append(scas)
        if scas == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((acas-scas)/scas)))
        
        scbc = int(row['銅槽B靶設定電流C'])
        acbc = int(row['銅槽B靶實際電流C'])
        data.append(scbc)
        if scbc == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((acbc-scbc)/scbc)))
        
        scbs = int(row['銅槽B靶設定電流S'])
        acbs = int(row['銅槽B靶實際電流S'])
        data.append(scbs)
        if scbs == 0:
            data.append(0)
        else:
            data.append(float('%.6f' % ((acbs-scbs)/scbs)))
        
        atime = []
        atime.append(str(datetime.datetime.strptime(row['生產開始時間'], "%Y/%m/%d %H:%M:%S")))
        atime.append(str(datetime.datetime.strptime(row['生產結束時間'], "%Y/%m/%d %H:%M:%S")))
        
        ttime = []
        ttime.append(str(datetime.datetime.strptime(row['錫槽A靶進入時間'], "%Y/%m/%d %H:%M:%S")))
        ttime.append(str(datetime.datetime.strptime(row['錫槽A靶離開時間'], "%Y/%m/%d %H:%M:%S")))
        
        ctime = []
        ctime.append(str(datetime.datetime.strptime(row['銅槽A靶進入時間'], "%Y/%m/%d %H:%M:%S")))
        ctime.append(str(datetime.datetime.strptime(row['銅槽A靶離開時間'], "%Y/%m/%d %H:%M:%S")))
        
        record[i]=(data,atime,ttime,ctime)
```

```python
deviceState = []
rf = open('deviceState_sorted.json', 'r')
for r in rf.readlines():
    dic = json.loads(r)
    for key in dic:
        deviceState.append((key,dic[key]))
```

```python
for r in record:
    acheck = 0
    tcheck = 0
    ccheck = 0
    for d in deviceState:
        if not d[1] and record[r][1][0] < d[0] < record[r][1][1]:
            acheck = 1
            if record[r][2][0] < d[0] < record[r][2][1]:
                tcheck = 1
            elif record[r][3][0] < d[0] < record[r][3][1]:
                ccheck = 1
            
    record[r] = (record[r][0], acheck, tcheck, ccheck)

print(record)
```

```python
jsObj = '\n'.join([json.dumps({r:record[r]}) for r in record])
fileObject = open('record.json', 'w')
fileObject.write(jsObj)
fileObject.close()
```
